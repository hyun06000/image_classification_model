```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

import tensorflow_datasets as tfds

from data.imagenet_labels import imagenet_labels
```

```python
#tfds.list_builders() #'imagenet2012'
ds_name = 'imagenet2012'
builder = tfds.builder(ds_name)
```

```python
builder.download_and_prepare()

'''
Downloading and preparing dataset Unknown size
(download: Unknown size, generated: 155.84 GiB, total: 155.84 GiB) to 
/root/tensorflow_datasets/imagenet2012/5.1.0...
Dataset imagenet2012 downloaded and prepared to 
/root/tensorflow_datasets/imagenet2012/5.1.0. 
Subsequent calls will reuse this data.
'''
```

```python
tr_ds, te_ds, val_ds = builder.as_dataset(split = ['train', 'test','validation'], shuffle_files = True)
print(tr_ds)
print(te_ds)
print(val_ds)
```

```python
indx, max_v, max_h = 0, 0, 0
for example in tr_ds:
    indx += 1
    im = example['image']
    max_v = max(max_v, im.shape[0])
    max_h = max(max_h, im.shape[1])

print('the_number_of_data : ', indx)
print('the_max_v_size : ', max_v)
print('the_max_h_size : ', max_h)



tr_ds = tr_ds.take(1)
for example in tr_ds:
    print(list(example.keys()))
    file_name = example['file_name']
    image = example['image']
    label = example['label']
    
    print('file_name : ',file_name)
    print(image.shape)
    plt.imshow(image)
    plt.show()
    print('label : ',imagenet_labels[label.numpy()])
```

```python
indx, max_v, max_h = 0, 0, 0
for example in val_ds:
    indx += 1
    im = example['image']
    max_v = max(max_v, im.shape[0])
    max_h = max(max_h, im.shape[1])

print('the_number_of_data : ', indx)
print('the_max_v_size : ', max_v)
print('the_max_h_size : ', max_h)


val_ds = val_ds.take(1)
for example in val_ds:
    print(list(example.keys()))
    file_name = example['file_name']
    image = example['image']
    label = example['label']
    
    print('file_name : ',file_name)
    plt.imshow(image)
    plt.show()
    print('label : ',imagenet_labels[label.numpy()])
```
